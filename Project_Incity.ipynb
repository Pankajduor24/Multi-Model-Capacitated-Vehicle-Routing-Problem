```python
"""Capacited Vehicles Routing Problem (CVRP)."""
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ortools.constraint_solver import routing_enums_pb2
from ortools.constraint_solver import pywrapcp


def capacity():
    ton = 907
    capacities = {'Tata Ace': {'weight': 0.85*ton, 'volume': 161, 'dist': 100}, 
                  'Mahindra Bolero Pickup': {'weight': 1.5*ton, 'volume':184, 'dist': 250}, 
                  'Elcher 14 Feet': {'weight': 4*ton, 'volume': 546, 'dist':400}, 
                  'Elcher 17 Feet': {'weight': 5*ton, 'volume': 714, 'dist': 50000},
                  'Elcher 20 Feet': {'weight': 7*ton, 'volume': 1120, 'dist': 50000}}
    default_utilization = [0.5, 0.5, 0.75, 0.75, 0.75, 0.75]
    check = input("Do you want to change the vehicle Usable CFT ratio? (y/n): ")
    if check.upper() == 'N':
        for index, vehicle in enumerate(capacities.keys()):
            capacities[vehicle]['volume'] = capacities[vehicle]['volume'] * default_utilization[index]
    else:
        for vehicle in capacities.keys():
            utilization = float(input(f"Enter Usable CFT ratio for {vehicle}: "))
            capacities[vehicle]['volume'] = capacities[vehicle]['volume'] * utilization
    return capacities

def haversine(lat1, lon1, lat2, lon2):
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat / 2) ** 2 + math.cos(lat1) * math.cos(lat2) * (math.sin(dlon / 2) ** 2)
    c = 2 * math.asin(math.sqrt(a))

    # Radius of Earth in kilometers
    radius_earth = 6371.0*1.25  # kilometers

    # Calculate the distance
    distance = radius_earth * c
    return distance

def distance_matrix(points):
    # Initialize distance matrix
    num_points = len(points)
    distance_matrix = np.zeros((num_points, num_points)).tolist()
    
    # Calculate distances between each pair of points
    for i in range(num_points):
        for j in range(num_points):
            distance_matrix[i][j] = math.ceil(haversine(points[i][0], points[i][1], points[j][0], points[j][1]))
    return distance_matrix

def get_vehicle_type(weight, volume, distance):
    """Returns the vehicle type based on weight and volume."""
    for vehicle, capacity in capacities.items():
        if (capacity['weight'] >= weight and capacity['volume'] >= volume) and capacity['dist'] >= distance:
            return vehicle
    return "No suitable vehicle found"

def draw_graph(tourss):
    # Plot each tour
    count = 0
    for tour in tourss:
        lats, lons, details, locations = zip(*tour)
        if len(tour) == 2:  # Only warehouse and one location
            count += 1
            plt.plot(lons, lats, marker='o', markersize=10, color='blue', label=f'Route {count}')
        else:
            count += 1
            plt.plot(lons, lats, marker='o', label=f'Route {count}')

        # Plot arrows indicating direction
        for i in range(len(tour)-1):
            dx = tour[i+1][1] - tour[i][1]
            dy = tour[i+1][0] - tour[i][0]
            plt.arrow(tour[i][1], tour[i][0], dx, dy, head_width=0.002, head_length=0.002, color='red')

        # Annotate locations
        for i, (lat, lon, _, loc_name) in enumerate(zip(lats, lons, details, locations)):
            plt.text(lon, lat, loc_name, fontsize=8, ha='right')

    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Tour Map')
    plt.grid(True)
    plt.legend()
    plt.show()

def multi_depot_loop():
    # Specify maximum distance in a route
    max_distance = int(input('Enter maximum distance of a trip: '))
    for depot_index, depot in enumerate(depots):
        # Filter rows where both 'LAT' and 'LONG' match the depot coordinates
        data = dat[(dat['LAT'] == depot[0]) & (dat['LONG'] == depot[1])]
        # Extract necessary columns and convert to list of tuples
        data = [(row['Latitude'], row['Longitude'], 
                 {'QTY': row['Sum of QTY'], 'weight': row['Sum of Total Weight'], 'volume': row['Sum of Total CFT']}, 
                 row['Destination Pin']) for _, row in data.iterrows()]
        inner_data = []
        outer_data = []
        for index, point in enumerate(data):
            # Convert latitude and longitude values to float
            depot_lat = float(depot[0])
            depot_lon = float(depot[1])
            point_lat = float(point[0])
            point_lon = float(point[1])

            # Calculate distance using haversine function
            distance_to_depot = haversine(depot_lat, depot_lon, point_lat, point_lon)
            if distance_to_depot <= depot_distance:
                inner_data = split_nodes(inner_data, point)
        
        inner_points = [tuple((depot[0], depot[1], {'QTY': 0, 'weight': 0, 'volume': 0}, depot[2]))] + inner_data
        
        inter_city_vrp = routing_inner(inner_points, max_distance)
    print()
    test = input('Enter any key to exit: ')


def split_nodes(inner_data, point):
    max_capacity_weight = max(capacity['weight'] for capacity in capacities.values())
    max_capacity_volume = max(capacity['volume'] for capacity in capacities.values())
    count = 0
    while point[2]['weight'] > max_capacity_weight or point[2]['volume'] > max_capacity_volume:
        excess_weight = point[2]['weight']
        excess_volume = point[2]['volume']
        excess_qty = point[2]['QTY']
        count += 1
        # Split the excess demand into virtual nodes to fill capacity proportionally
        if max_capacity_weight > max_capacity_volume:
            virtual_qty_allowed = int(max_capacity_volume * excess_qty / excess_volume)
            if virtual_qty_allowed > 0:
                virtual_node_volume = virtual_qty_allowed * excess_volume / excess_qty
                virtual_node_weight = virtual_qty_allowed * excess_weight / excess_qty

                virtual_node = (point[0], point[1], 
                                {'weight': virtual_node_weight, 'volume': virtual_node_volume, 
                                 'QTY': virtual_qty_allowed}, f'{point[3]}-{count}')

                # Update remaining excess
                excess_weight -= virtual_node_weight
                excess_volume -= virtual_node_volume
                excess_qty -= virtual_qty_allowed

                # Update the original point with the remaining excess
                point = (point[0], point[1], 
                         {'weight': excess_weight, 'volume': excess_volume, 
                          'QTY': excess_qty}, point[3])

                # Append the virtual node and updated point to the inner_data list
                inner_data.append(virtual_node)
                continue
            else:
                inner_data.append(point)
                break
    inner_data.append(point)
        
    return inner_data

def create_data_model(nodes):
    """Stores the data for the problem."""
    max_weight = max([capacity['weight'] for capacity in capacities.values()])
    max_volume = max([capacity['volume'] for capacity in capacities.values()])
    data = {}
    data["distance_matrix"] = distance_matrix(nodes)
    data["demands_weight"] = [math.ceil(point[2]['weight']) for point in nodes]
    data["demands_volume"] = [math.ceil(point[2]['volume']) for point in nodes]
    data['demands'] = [point[2]['QTY'] for point in nodes]
    data["num_vehicles"] = len(nodes) #int((sum(data['demands_volume'])) / (max_volume)) + 2
    data["vehicle_capacities_weight"] = [math.ceil(max_weight) for _ in range(data['num_vehicles'])]
    data["vehicle_capacities_volume"] = [math.ceil(max_volume) for _ in range(data['num_vehicles'])]
    data["depot"] = 0
    return data

def print_solution(data, manager, routing, solution, nodes):
    demand_weight = data["demands_weight"]
    demand_volume = data["demands_volume"]
    total_load_weight = sum(demand_weight)
    total_load_volume = sum(demand_volume)
    """Prints solution on console."""
    if total_load_weight > 0 or total_load_volume > 0:
        print(f'\n\nWarehouse: {nodes[0][3]}\n')
        # print(f" Objective: {solution.ObjectiveValue()}")
        total_distance = 0
        total_volume_load = 0
        total_weight_load = 0
        total_nodes = 0
        tours = []
        v_count = 0
        for vehicle_id in range(data["num_vehicles"]):
            index = routing.Start(vehicle_id)
            tours.append([])
            plan_output = ""
            route_distance = 0
            route_load_weight = 0
            route_load_volume = 0
            num_nodes = 0
            while not routing.IsEnd(index):
                num_nodes += 1
                node_index = manager.IndexToNode(index)
                route_load_weight += demand_weight[node_index]
                route_load_volume += demand_volume[node_index]
                plan_output += f" {nodes[node_index][3]} Load(W:{demand_weight[node_index]}, V:{demand_volume[node_index]}) -> "
                previous_index = index
                index = solution.Value(routing.NextVar(index))
                route_distance += routing.GetArcCostForVehicle(
                    previous_index, index, vehicle_id
                )
                tours[-1].append(nodes[node_index])
            else:
                node_index = manager.IndexToNode(index)
                tours[-1].append(nodes[node_index])
            plan_output += f" {nodes[manager.IndexToNode(index)][3]} Load(W:{demand_weight[manager.IndexToNode(index)]}, V:{demand_volume[manager.IndexToNode(index)]})\n"
            plan_output += f"Distance of the route: {route_distance}KM\n"
            plan_output += f"Load of the route: Weight({route_load_weight}), Volume({route_load_volume})\n"
            if route_load_weight > 0 or route_load_volume > 0:
                v_count += 1
                print(f'Route for vehicle {v_count}:\n', plan_output)
                print(f'Customer locations covered: {num_nodes - 1}')
                vehicle_type = get_vehicle_type(route_load_weight, route_load_volume, route_distance)
                print("Suitable vehicle: ", vehicle_type)
                print()
            total_distance += route_distance
            total_volume_load += route_load_volume
            total_weight_load += route_load_weight
            total_nodes += num_nodes - 1
        print(f"Total distance of all routes: {total_distance}KM")
        print(f"Total load of all routes: (W: {total_weight_load}, V: {total_volume_load})")
        print(f'Total customers served: {total_nodes}')
        tourss = []
        for tour in tours:
            if len(tour)>2:
                tourss.append(tour)
        draw_graph(tourss)
        
def routing_inner(nodes, max_distance):
    """Solve the CVRP problem."""
    # Instantiate the data problem.
    data = create_data_model(nodes)
    
    # Create the routing index manager.
    manager = pywrapcp.RoutingIndexManager(
        len(data["distance_matrix"]), data["num_vehicles"], data["depot"]
        )

    # Create Routing Model.
    routing = pywrapcp.RoutingModel(manager)


    def distance_callback(from_index, to_index):
        """Returns the distance between the two nodes."""
        # Convert from routing variable Index to distance matrix NodeIndex.
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return data["distance_matrix"][from_node][to_node]

    transit_callback_index = routing.RegisterTransitCallback(distance_callback)

    # Define cost of each arc.
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

    # Add Capacity constraint.
    def demand_callback_weight(from_index):
        """Returns the demand of the node."""
        # Convert from routing variable Index to demands NodeIndex.
        from_node = manager.IndexToNode(from_index)
        return data["demands_weight"][from_node]

    demand_callback_index_weight = routing.RegisterUnaryTransitCallback(demand_callback_weight)

    # Add Capacity constraint.
    def demand_callback_volume(from_index):
        """Returns the demand of the node."""
        # Convert from routing variable Index to demands NodeIndex.
        from_node = manager.IndexToNode(from_index)
        return data["demands_volume"][from_node]

    demand_callback_index_volume = routing.RegisterUnaryTransitCallback(demand_callback_volume)
    
    routing.AddDimensionWithVehicleCapacity(
        demand_callback_index_weight,
        0,  # null capacity slack
        data["vehicle_capacities_weight"],  # vehicle maximum capacities
        True,  # start cumul to zero
        "Capacity_weight",
    )
    
    routing.AddDimensionWithVehicleCapacity(
        demand_callback_index_volume,
        0,  # null capacity slack
        data["vehicle_capacities_volume"],  # vehicle maximum capacities
        True,  # start cumul to zero
        "Capacity_volume",
    )

    # Add distance constraint
    distance_dimension_name = "Distance"
    routing.AddDimension(
        transit_callback_index,
        0,  # no slack
        max_distance,  # vehicle maximum travel distance
        True,  # start cumul to zero
        distance_dimension_name,
    )
    distance_dimension = routing.GetDimensionOrDie(distance_dimension_name)
    
    # Setting first solution heuristic.
    search_parameters = pywrapcp.DefaultRoutingSearchParameters()
    search_parameters.first_solution_strategy = (
        routing_enums_pb2.FirstSolutionStrategy.GLOBAL_CHEAPEST_ARC
    )
    search_parameters.local_search_metaheuristic = (
        routing_enums_pb2.LocalSearchMetaheuristic.GUIDED_LOCAL_SEARCH
    )
    search_parameters.time_limit.FromSeconds(1)

    # Solve the problem.
    solution = routing.SolveWithParameters(search_parameters)

    # Print solution on console.
    if solution:
        print_solution(data, manager, routing, solution, nodes)
    else:
        print(f"\n\nNo solution found for warehouse: {nodes[0][3]} !")


file_name = input('Enter data file name (.xlsx): ')
sheet_name = input('Enter sheet name: ')

print('Please make sure that your data file should be in the same folder as your program file!')

# Read the Excel file
df = pd.read_excel(f'{file_name}.xlsx', sheet_name=sheet_name)

# Assuming the header row is the third row
header = df.iloc[2]

# Skip the first 3 rows and reset index
df = df[3:].reset_index(drop=True)

# Set the columns to the header row
df.columns = header

# Drop rows with NaN values in 'LAT', 'LONG', 'Latitude', or 'Longitude' columns
df = df.dropna(subset=['LAT', 'LONG', 'Latitude', 'Longitude'])

# Convert 'Month' column to proper date format
df['Month'] = pd.to_datetime(df['Month'], format='%d-%m-%Y')

date = input('Enter date for which you want to perform the task (YYYY-MM-DD): ')

# Filter data for the specific month
dat = df[(df['Month'] == date)]

# Capacities for each type of vehicle
capacities = capacity()

# Depot coordinates
m_depot = dat[['LAT', 'LONG', 'Region']].drop_duplicates()
depots = [tuple(row) for row in m_depot.values]

# Specify maximum distance from depot
depot_distance = 60  # Adjust as needed


# Call the function
multi_depot_loop()
```
